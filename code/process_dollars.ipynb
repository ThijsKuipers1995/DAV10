```python
df = pd.read_csv("../Week1/global_data_processing /average_product_prices_per_country_per_month.csv")
rates = pd.read_csv("../Week1/global_data_processing /exchange_rates_to_us_dollar.csv")

product = 'Wheat'
year = 2005
```

```python
def monthly_product_price_countries(product, year, month, df, rates):
    avg_prices = []
    df = df[(df['product']==product) & (df['year']==year) & (df['month']==month)]
    countries = sorted(list(set(df['country'])))
    for country in countries: 
        df_country = df[df['country']==country]
        prices = list(df_country['price_in_dollars'])
        avg_price = sum([float(price) if price != 'UNKNOWN' else 0 for price in prices])
        avg_prices.append(avg_price)
        print(avg_price)
    print(len(avg_prices))
    print(len(countries))
    
monthly_product_price_countries('Bread', 2014, 1, df, rates)
```

```python
#print(rates)
#print(df)
def convert_rates(df, rates):
    conv_rates = []
    for row in df.iterrows():
        product = row[1]['country']
        currency = row[1]['currency']
        month = row[1]['month']
        year = row[1]['year']
        avg_price = row[1]['avg_price']
        rateSeries = rates.loc[(rates['year']==year) & (rates['month']==month) & (rates['currency']==currency)]['value']
        if len(rateSeries) == 1 and avg_price != 'UNKNOWN':
            rate = list(rateSeries)[0]
            conv_rates.append(float(avg_price) / float(rate))
        else: conv_rates.append('UNKNOWN')
    return conv_rates
```

```python
r = convert_rates(df, rates)
print(len(r))
```

```python
df['price_in_dollars'] = r
```
